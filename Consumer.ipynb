```python
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
from s3fs import S3FileSystem
```

```python
consumer = KafkaConsumer(
    'testTopic2',
     bootstrap_servers=['18.189.29.102:9092'], 
    value_deserializer=lambda x: loads(x.decode('utf-8')))
```

```python
for i in consumer:
    print(i.value)
```

```python
s3 = S3FileSystem()
```

```python
for count, i in enumerate(consumer):
#     print(count)
#     print(i.value)
    with s3.open("s3://stock-market-kafka-mabuomar/stock_market_{}.json".format(count), 'w') as file:
        json.dump(i.value, file)  
```
